### Checking for any inconsistences on our data

#### Seconds
When we observe our boxplot for seconds, we notice a huge outlier that affects the entire distribution of our data

```python
# checking for rides greater than 6 hours (21600 seconds)
seconds_outlier = len(df_new[df_new['second'] > 21600])
total_rows = len(df_new)
pct_outlier = seconds_outlier / total_rows
outlier_booking_ids = df_new[df_new['second'] > 21600]['bookingID'].unique()
print(f'Total Number of rows with anomalies {seconds_outlier}')
print(f'Percentage of rows with anomalous seconds: {round(pct_outlier * 100,8)}%')
print(f'Anomalous Values: {df_new[df_new["second"] > 21600]["second"].unique()}')
print(f'BookingIDs with anomalous seconds: {outlier_booking_ids}')
```

It is practically impossible for rides to be 173,032 Hrs long. This anomaly may be attributed to errors within the sensor of the taxi as there is multiple of the same second anomaly in different bookingIDs.

## Statistical Tests
We will be doing 2 statisical tests
- Point Biserial Correlation between label and numerical columns
- Chi-squared Test for Correlation between label and nominal variables (eg. car brand)

### Point Biserial Correlation

```python
for col in continuous_cols:
    correlation = stats.pointbiserialr(df_new["label"], df_new[col]).statistic
    print(f'\nCorrelation between label and {col}:')
    print(f'{format(correlation, ".7f")}')
```

We can see that all of our continous variables have weak linear relationships with the safety of the ride

### Chi-squared test

```python
label = np.array(df_new["label"]).reshape(-1, 1)
car_brand = np.array(df_new["car_brand"]).reshape(-1, 1)

chi2_stat, p_values = chi2(label, car_brand)

print(f'Chi-Square Test for {df_new["car_brand"].name}\n p-value: {p_values[0]}')
```

## Amount of dangerous rides by age

```python
unqiue_bookingID = df_new.groupby('bookingID').first()
rides_by_age = unqiue_bookingID.groupby(['age', 'label'])['label'].count().unstack().fillna(0)
rides_by_age = rides_by_age.sort_values(1, ascending=False)

# Calculate the percentage of dangerous rides for each age group
rides_percentage = rides_by_age[1] / rides_by_age.sum(axis=1) * 100

fig, ax = plt.subplots(figsize=(15, 8))
bars = rides_by_age.plot(kind='bar', stacked=True, color=[palette[-1], palette[0]], ax=ax)

for i, age in enumerate(rides_by_age.index):
    percentage = rides_percentage.loc[age]
    ax.text(i, rides_by_age.loc[age].sum(), f'{percentage:.1f}%', ha='center', va='bottom')

for spine in ax.spines.values():
    spine.set_visible(False)

ax.set_yticks([])

ax.set_xlabel('Age')
ax.set_ylabel('Number of Rides')
ax.set_title('Number of Rides by Age', fontsize=20)
ax.legend(title='Label', labels=['Safe', 'Dangerous'])
plt.show()
```

## Dangerous Rides by years of experience

```python
bins = [5, 10, 15, 20, 25, 30]
labels = ['6-10', '11-15', '16-20', '21-24', '25+']

df_new['exp_group'] = pd.cut(df_new['years_of_exp'], bins=bins, labels=labels, right=False)

dangerous_rides_by_exp = df_new[df_new['label'] == 1].groupby('exp_group').size()
total_rides = df_new['label'].count()
rides_percentage = dangerous_rides_by_exp / total_rides * 100

# Plot the bar graph
fig, ax = plt.subplots(figsize=(15, 8))
bars = ax.barh(dangerous_rides_by_exp.index, dangerous_rides_by_exp.values, color='red')

for spine in ax.spines.values():
    spine.set_visible(False)

ax.set_xticks([])

# Add the percentages on top of the bars
for i, bar in enumerate(bars):
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height() / 2,
            f'{rides_percentage[i]:.1f}%', ha='left', va='center')

ax.set_ylabel('Years of Experience', fontsize=12)
ax.set_title('Number of Dangerous Rides by Years of Experience', fontsize=18)
plt.show()
```

We can see that inexperienced drivers tend to have higher amounts of dangerous rides.

## Most Dangerous drivers in the company

```python
unqiue_bookingID = df_new.groupby('bookingID').first()
dangerous_rides_by_booking = unqiue_bookingID[unqiue_bookingID['label'] == 1].groupby('bookingID')['label'].first()

dangerous_rides_by_driver = unqiue_bookingID[unqiue_bookingID['label'] == 1].groupby('driver_name')['label'].size()

total_rides_by_driver = unqiue_bookingID.groupby('driver_name')['label'].count()

# Calculate the percentage of dangerous rides for each driver
dangerous_rides_by_driver_percentage = dangerous_rides_by_driver / total_rides_by_driver * 100

# Get the top 10 most dangerous drivers
top_dangerous_drivers_percentage = dangerous_rides_by_driver_percentage.sort_values(ascending=False).head(10)

fig, ax = plt.subplots(figsize=(15, 8))
colors = [palette[0] if i < 3 else palette[3] for i in range(len(top_dangerous_drivers_percentage))]

bars = ax.bar(top_dangerous_drivers_percentage.index, top_dangerous_drivers_percentage.values, color=colors)

# Add the percentages on top of the bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height,
            f'{top_dangerous_drivers_percentage.values[i]:.1f}%', ha='center', va='bottom')
    
for spine in ax.spines.values():
    spine.set_visible(False)

ax.set_yticks([])

ax.set_xlabel('Driver Name')
ax.set_ylabel('Percentage of Dangerous Rides')
ax.set_title('Top 10 Most Dangerous Drivers (Percentage)', fontsize=18)
ax.set_xticklabels(top_dangerous_drivers_percentage.index, rotation=45)
plt.show()
```

## Percentage of High risk rides

```python

average_speed_acceleration = df_new.groupby('bookingID').agg({
    'speed': 'mean',
    'acceleration_x': 'mean',
    'label': 'first'
})
average_speed_acceleration.drop(average_speed_acceleration[average_speed_acceleration['label'] == 0].index, inplace=True)
## classifying high risk rides
def classify_risk(speed, acceleration):
    if speed > threshold_speed and acceleration > threshold_acceleration:
        return 'High Risk'
    elif speed > threshold_speed or acceleration > threshold_acceleration:
        return 'Medium Risk'
    else:
        return 'Low Risk'

threshold_speed = 10  
threshold_acceleration = 0.5  

average_speed_acceleration['risk_category'] = average_speed_acceleration.apply(lambda row: classify_risk(row['speed'], row['acceleration_x']), axis=1)

risk_counts = average_speed_acceleration['risk_category'].value_counts()

fig, ax = plt.subplots(figsize=(15, 8))
explode = (0, 0, 0.1)

ax.pie(risk_counts, labels=risk_counts.index, autopct='%1.1f%%', startangle=90, colors=[palette[-1], palette[-2], palette[0]], explode=explode, shadow=True)

ax.set_title('Distribution of Risk Categories', fontsize=18)

ax.legend(risk_counts.index, title="Risk Categories", loc="upper right")

plt.show()
```

We can see that most dangerous rides are in the low risk category, with the average speed and average acceleration being lower than 10 and 0.5 respectively. 5.6% of dangrous rides have average speed and accerleration ebing higher than our threshold.

## Dangerous Rides by Gender

```python
unqiue_bookingID = df_new.groupby('bookingID').first()
dangerous_rides_by_gender = unqiue_bookingID.groupby(['gender', 'label'])['label'].count().unstack()

total_rides_by_gender = unqiue_bookingID.groupby('gender')['label'].count()

# Calculate the percentage of dangerous and non-dangerous rides for each gender
dangerous_rides_by_gender_percentage = dangerous_rides_by_gender.divide(total_rides_by_gender, axis=0) * 100

fig, ax = plt.subplots(figsize=(12, 7))
bars = dangerous_rides_by_gender_percentage.plot(kind='bar', stacked=True, color=[palette[-1], palette[0]], ax=ax)

for i, (gender, row) in enumerate(dangerous_rides_by_gender_percentage.iterrows()):
    cum_height = 0
    for j, col in enumerate(row):
        ax.text(i, cum_height + col/2, f'{col:.1f}%', ha='center', va='bottom')
        cum_height += col

for spine in ax.spines.values():
    spine.set_visible(False)

ax.set_yticks([])

ax.set_xlabel('Gender', fontsize=12)
ax.set_ylabel('Percentage of Rides', fontsize=12)
ax.set_title('Percentage of Dangerous and Non-Dangerous Rides by Gender', fontsize=18)
ax.legend(title='Label', labels=['Safe', 'Dangerous'])
ax.set_xticklabels(labels=dangerous_rides_by_gender.index, rotation=0)
plt.show()
```

We can see that there is no difference between either gender, when it comes to dagnerous driving. The percentages of dangerous rides in respect to the total rides by gender are almost the same.
